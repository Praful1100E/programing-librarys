```python
import matplotlib.pyplot as plt

bins = [15, 20, 25, 30, 35, 40]
datascience_student_age = [43, 43, 40, 12, 65, 20]
js_student_age = [32, 21, 43, 56, 76, 80]

plt.hist(datascience_student_age, bins=bins, rwidth=0.8, histtype='bar',
         orientation='vertical', color='r', label='DataScience')

plt.hist(js_student_age, bins=bins, rwidth=0.8, histtype='bar',
         orientation='vertical', color='b', label='Javascript')

plt.title("Student")  #  Corrected: remove = sign
plt.legend(loc=3)     #  loc=3 means 'lower left'
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.show()
```

```python
from matplotlib import style
import matplotlib.pyplot as plt

style.use('ggplot')  # Apply ggplot style

datascience_student_age = [43, 43, 23, 12, 65, 89]
javascript_student_age = [32, 21, 43, 56, 76, 89]
bins = [15, 20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 90]

plt.figure(figsize=(16, 9))  # Set figure size

plt.hist(
    [datascience_student_age, javascript_student_age],  # Multiple datasets
    bins=bins,
    rwidth=0.8,  # Bar width
    orientation='vertical',  # Correct spelling
    color=['r', 'b'],  # Colors for each dataset
    label=['DataScience', 'JavaScript']  # Labels for legend
)

plt.title("Student Age Distribution")
plt.xlabel("Student Age Category")
plt.ylabel("Number of Students")
plt.legend(loc=0)  # Correct spelling
plt.show()
```
