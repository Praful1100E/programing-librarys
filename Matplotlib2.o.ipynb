```python
import matplotlib.pyplot as plt

# Student data
students = ['Amit', 'Riya', 'John', 'Neha', 'Raj']
marks = [85, 92, 76, 88, 95]

# Create the bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(students, marks, color='skyblue')

# Add title and labels
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

# Display marks on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, str(yval), ha='center', va='bottom')

# Show grid and plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 100)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Student data
students = ['Amit', 'Riya', 'John', 'Neha', 'Raj']
marks = [85, 92, 76, 88, 95]

# Create the bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(students, marks, color='skyblue')

# Add title and labels
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

# Add marks above bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, str(yval), ha='center', va='bottom')

# Add Y-axis grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Set Y-axis limit
plt.ylim(0, 100)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['axes.facecolor'] = 'honeydew'   # light green background
plt.rcParams['axes.edgecolor'] = 'green'
plt.rcParams['xtick.color'] = 'green'
plt.rcParams['ytick.color'] = 'green'
plt.rcParams['text.color'] = 'darkgreen'

# Plot example
x = [1, 2, 3, 4]
y = [10, 20, 15, 30]

plt.plot(x, y, color='green', marker='o')
plt.title("Custom Green Style")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

# Generate random student names
students = ['Amit', 'Riya', 'John', 'Neha', 'Raj', 'Simran', 'Aman', 'Tina']
# Generate random marks between 50 and 100 for each student
marks = [random.randint(50, 100) for _ in students]

# Plotting
plt.figure(figsize=(10, 5))
bars = plt.bar(students, marks, color='green')

# Title and labels
plt.title("Random Student Marks (Green Theme)")
plt.xlabel("Students")
plt.ylabel("Marks")

# Display marks above each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, str(yval), ha='center', va='bottom', color='black')

# Grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Limit Y-axis
plt.ylim(0, 110)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

# Generate 100 random marks between 40 and 100
marks = [random.randint(40, 100) for _ in range(100)]

# Create histogram
plt.figure(figsize=(10, 5))
plt.hist(marks, bins=[40, 50, 60, 70, 80, 90, 100], color='green', edgecolor='black')

# Add title and labels
plt.title("Histogram of Random Student Marks")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")

# Add grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
